### The scatter plot shows that from between the age 25 and 35 on average that age range is doing more sit ups then the other ages.

```python
fig= plt.figure(figsize=(10,10))
scatter_plot=sns.scatterplot(x='age',
                             y='sit-ups counts',
                             data=df,
                             color='red')
```

### Ignoring the two outliers, which I belive to be mistakes in the entry of the data, we can see that age doesn't really affect the sit and bend forward ability between the different ages.

```python
fig= plt.figure(figsize=(10,10))
scatter_plot=sns.scatterplot(x='age',
                             y='sit and bend forward_cm',
                             data=df,
                             color='blue')
```

### Looking at this scatter plot, we can examine that the younger the age the higher the broad jump. As the age progresses the distances decrease.

```python
fig= plt.figure(figsize=(10,10))
scatter_plot=sns.scatterplot(x='age',
                             y='broad jump_cm',
                             data=df,
                             color='blue')
```

### Looking at this pairplot we can see all the columns in our dataset compared to one another, though specific conclusions are hard to be made from this, in general we can see that for the most part besides for outliers, there is correlation that can be found between them.

```python
# let's look and examine the correlations between the columns
df.corr()
```

```python
# This is a correlation matrix for all the continuous variables
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# This is the correlation matrix displyed in a figure
fig = plt.figure(figsize=(8,8))
sns.heatmap(df.corr(method='pearson'),annot=True,fmt='.2f',mask=np.triu(df.corr(method='pearson')),)
```

## Data Set Graphical Exploration

```python
# age vs body fat % to see if there is any visible pattern

plt.bar(df['age'], df['body fat_%'], color='red')
plt.xlabel('Age')
plt.ylabel('Body Fat%')
plt.title('Age vs Body Fat%')
```

```python
# weight vs gripforce to see how it compares to one another

joint_plot=sns.jointplot(x='height_cm',
                         y='gripForce',
                        data=df,
                        kind='reg',
                        color='green')
```

```python
# weight vs gripforce to see how it compares to one another

joint_plot=sns.jointplot(x='weight_kg',
                         y='gripForce',
                        data=df,
                        kind='reg',
                        color='blue')
```

```python
# weight vs systolic to see how it compares to one another

plt.scatter(df['weight_kg'],df['systolic'])
plt.xlabel('Weight (KG)')
plt.ylabel('Systolic')
plt.show()
```

```python
# weight vs diastolic to see how it compares to one another
plt.scatter(df['weight_kg'],df['diastolic'], color='green')
plt.xlabel('Weight (KG)')
plt.ylabel('Diastolic')
plt.show()
```
